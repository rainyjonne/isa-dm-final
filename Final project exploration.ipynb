### Situation1:  test with player "justin tucker" & "Ray-Ray Armstrong" for their speed trend growth

```python
justin_df = track_df.loc[track_df["displayName"] == "Justin Tucker"]
```

```python
justin_df
```

```python
# see for match in 2018090900 & play id is 380
justin_df.loc[(justin_df["gameId"] == 2018090900) & (justin_df["playId"] == 380)]
```

```python
# see for match in 2018090900 & play id is 400
justin_df.loc[(justin_df["gameId"] == 2018090900) & (justin_df["playId"] == 400)]
```

```python
# generate ray ray df
ray_df = track_df.loc[track_df["displayName"] == "Ray-Ray Armstrong"]
test_df = pd.concat([justin_df, ray_df])
test_df
```

```python
# sum up the distance he ran in each game
dis_df = test_df.groupby(["gameId", "displayName"]).sum()['dis'].reset_index()
dis_df
```

```python
# count frames he was recorded in each game
frame_df = test_df.groupby(["gameId", "displayName"]).count()['frameId'].reset_index()
frame_df = frame_df.rename(columns={'frameId': 'frame_counts'})
frame_df
```

```python
# get an average speed per frame dataframe
avg_df = dis_df.merge(frame_df, on=["gameId", "displayName"])

avg_df['avg_frame_speed'] = avg_df['dis'] / avg_df['frame_counts']
avg_df['gameId'] = avg_df.astype({"gameId": str})
avg_df['gameId'] = pd.to_datetime(avg_df['gameId'], format="%Y%m%d%H")
avg_df.sort_values("gameId").reset_index(drop=True)
```

```python
avg_df
```

```python
# Answer here
import plotly.express as px
fig = px.line(avg_df, x=avg_df["gameId"], y="avg_frame_speed", color="displayName")
fig.update_traces(mode='markers+lines')
fig.update_layout(
    xaxis_type = 'date'
)
```

```python
# lets see the special mentions time comparisons among two kickers (just for a demo)
# lets pick "Justin Tucker" & "Jason Sanders"

j_perf_df = play_df[play_df["playDescription"].str.contains("J.Tucker")]
j_games = len(j_perf_df.groupby("gameId").count())
j_mentions = len(j_perf_df)
```

```python
j2_perf_df = play_df[play_df["playDescription"].str.contains("J.Sanders")]
j2_games = len(j2_perf_df.groupby("gameId").count())
j2_mentions = len(j2_perf_df)
```

```python
perf_check_df = pd.DataFrame(data={"game_times": [j_games, j2_games], "mentions": [j_mentions, j2_mentions]})
perf_check_df["pergame_mentions"] = perf_check_df['mentions'] / perf_check_df['game_times']
perf_check_df
```

#### You can see from the times that J.Tucker has been mentioned for about 10 times per game while J.Sanders has been mentioned for about 8 times per game.
